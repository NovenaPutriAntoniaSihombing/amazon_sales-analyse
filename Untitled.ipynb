```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
data=pd.read_csv("/Users/udaysinghshekhawat/Documents/amazon_sales/amazon.csv")
data.head()
```

```python
data.info()
```

Exploratory Data Analysis (EDA)

```python
# Check for missing values
missing_values = data.isnull().sum()
print(missing_values)
```

```python
# Descriptive statistics for numerical columns
print(data.describe())
```

```python
# Top 5 products with the highest discounts
top_discounted_products = data[['product_name', 'discount_percentage']].sort_values(by='discount_percentage', ascending=False).head(5)
print(top_discounted_products)
```

```python
# Correlation between discount percentage and rating
plt.figure(figsize=(8,6))
sns.scatterplot(x='discount_percentage', y='rating', data=data)
plt.title('Discount Percentage vs Rating')
plt.show()
```

Analyzing Product Categories

```python
# Count of products per category
plt.figure(figsize=(11,6))
sns.countplot(y='category', data=data, order=data['category'].value_counts().index)
plt.title('Number of Products per Category')
plt.show()
```

User Analysis

```python
# Users with the most reviews
top_reviewers = data['user_name'].value_counts().head(1)
print(top_reviewers)
```

```python
# Plot the distribution of product ratings
plt.figure(figsize=(8,6))
sns.histplot(data['rating'], bins=10, kde=True)
plt.title('Distribution of Product Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot the distribution of discount percentages
plt.figure(figsize=(8,6))
sns.histplot(data['discount_percentage'], bins=15, kde=True)
plt.title('Distribution of Discount Percentages')
plt.xlabel('Discount Percentage')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter plot for Rating vs. Rating Count
plt.figure(figsize=(8,6))
sns.scatterplot(x='rating', y='rating_count', data=data)
plt.title('Rating vs. Rating Count')
plt.xlabel('Rating')
plt.ylabel('Number of Reviews')
plt.show()
```

```python
# Plot the distribution of actual product prices
plt.figure(figsize=(8,6))
sns.histplot(data['actual_price'], bins=20, kde=True)
plt.title('Distribution of Actual Product Prices')
plt.xlabel('Actual Price')
plt.ylabel('Frequency')
plt.show()

```

```python
# Jointplot for discount percentage vs. rating count
sns.jointplot(x='discount_percentage', y='rating_count', data=data, kind='scatter', height=8)
plt.title('Discount Percentage vs Rating Count')
plt.show()
```
